# Projections and Coordinate Systems

We can further check the coordinate system of the map (states_limited) we just generated.

```python
print(states_limited.crs)
```

According to lessons/03_crs_map_projections.ipynb in D-Lab Python Geospatial Fundamentals Workshop, 4326 stands for WGS84 and 3310 stands for Albers Equal Area projection.

The following map contains the same geographical information as the states_limited map above but with a different projection and coordinate system.

```python
states_limited.crs = None
states_limited.crs = "epsg:3310"
states_limited.crs
states_limited.plot(color='blue');
```

For comparison, I change the coordinate system back to 4326. With a different projection, the size of US territory is changed and this shows that the difference in the projections and coordinate systems used in mapping demonstrates different perspectives. The mapper's decision in choosing which projection makes the map convey an intended perspective of the place.

```python
states_limited.crs = None
states_limited.crs = "epsg:4326"
states_limited.crs
states_limited.plot(color='blue');
```

The following maps are further customized in color, X, Y axes, and figure size.

```python
fig, ax = plt.subplots(figsize=(10, 10))

states_limited.plot(color='purple', ax=ax)

states_limited.plot(color='lightgrey', ax=ax)

ax.set_xlim(-140, -50)
ax.set_ylim(20, 50)
```

# States in Details and Aesthetic of Maps
In this section, I elaborate on the ways in which maps are political due to the decisions regarding the size of the image, specificity of information, and classification colors with examples of Hawaii and Washington.

We pick out the information regarding Hawaii with the STATEFP 15. The file here is still "cb_2018_us_county_5m.zip" in the Data folder in DIGHUM101-2023.

```python
hi_counties = counties.loc[counties["STATEFP"] == "15"]
print(hi_counties.shape)
hi_counties.head()
```

This is a map with only the shape of Hawaii.

```python
hi_counties.plot();
```

However, if we generate the map with state boundaries with no further customizations, Hawaii would look smaller shown as a state.

```python
state_boundaries
```

```python
hawaii = state_boundaries.loc[state_boundaries["ABBREV"] == "HI"]
hawaii
```

```python
hawaii.plot(); 
```

The ways in which different settings in plotting generate different ways of visualizing the same space convey different focuses of the map. While the first map of Hawaii may emphasize the shape of the islands and centers the shapes, the second map shows Hawaii on the side. The size of a represented space could be a way of demonstrating its importance on the map, the same as if the space is placed in the center of the map. 

The next case study is the state of Washington. The first step is to pick out the shapes of counties. The second is for the shape of the entire state.

```python
wa_counties = counties.loc[counties["STATEFP"] == "53"]
print(wa_counties.shape)
wa_counties.head()
```

```python
washington = state_boundaries.loc[state_boundaries["ABBREV"] == "WA"]
washington
```

The first map I generate still only contains the shape of the state.

```python
washington.plot(color = "pink"); 
```

The second map is still the map of Washington but with county polygons. 

```python
wa_counties.plot();
```

This step is to prepare for further customization of the map of Washington.

```python
pop = wa_counties
pop.head()
```

The following map is Washington with state polygons and counties with further color classifications to show shapes of counties.

```python
pop.plot(
    
    cmap = "Blues", 
    
    edgecolor = "pink", 
    
    linewidth = 0.5, 
    
    figsize = (10, 10),
    
    legend = True,
    
    k = 10
);
```

I then test the map with a different color and figure size.

```python
pop.plot(
             cmap = "Purples", 
             edgecolor = "pink", 
             linewidth = 0.1,  
             legend = True, 
             figsize = (15,12),
             k = 7   
            );
```

I decided that I still prefer the blue version but I do like the 15, 12 figure size so I ended up using the blue color but with the 15, 12 figure size. I also removed the axes and added a title, as well as names of counties.
This final step intends to show that the map is produced through the mapper's choices and that the names and titles allow the map to cater to a wider population. The map that only shows the shape of Washington may still be recognizable to some people, but with a title, the polygons of the counties, and the names of the counties would make the map have more directory functions and serve more purposes. I believe the findings in this research should not be interpreted as which map is better but they show that maps are made with choices so that they are not neutral nor the reality, they are simply representations of the world, the same as data.

Part of the following commands is referenced from https://towardsdatascience.com/lets-make-a-map-using-geopandas-pandas-and-matplotlib-to-make-a-chloropleth-map-dddc31c1983d, as indicated below. 

```python
import matplotlib.pyplot as plt

pop_points = pop.to_crs('+proj=cea').centroid.to_crs(pop.crs)

ax = pop.plot(figsize = (15, 12), cmap = "Blues", 
                  edgecolor = "pink", linewidth = 0.5)

#The following two lines are referenced from https://towardsdatascience.com/lets-make-a-map-using-geopandas-pandas-and-matplotlib-to-make-a-chloropleth-map-dddc31c1983d

ax.axis("off")
ax.set_title("WA", fontdict={"fontsize": "25", "fontweight" : "3"})

texts = []
for x, y, label in zip(pop_points.geometry.x, pop_points.geometry.y, pop["NAME"]):
    texts.append(plt.text(x, y, label, fontsize = 8))
    
```
